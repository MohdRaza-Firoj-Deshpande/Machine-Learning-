One-hot encoding is a process that converts categorical data into a numerical format for machine learning by creating a new binary column for each unique category. Foreach data point, the corresponding column gets a value of 1, while all other new columns for that category get a 0. This is a common pre- processing step for nominal variables that lack an inherent order.

For Eg. If we had
categorical variable color with 3 value Red Green and Bule they can be represented as

Red(1,0,0)
Green(0,1,0)
Bule(0,0,1)

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.DataFrame({'color':['Red','Blue','Green','Green','Blue'] })
```

```python
df.head()
```

```python
#Create a instance for OneHotEncoder
encoder=OneHotEncoder()
```

```python
#Perform Fit and Transform
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
#for new data
encoder.transform([['Blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```
